# Pandas, Part 2
loc and iloc are two fundamental data selection methods in Pandas DataFrames, differing in how they identify rows and columns:

loc (Label-based indexing):

Selects data based on the labels (names) of rows and columns.

Can use single labels, lists of labels, or slices of labels.

When slicing with loc, both the start and end labels are inclusive.

Supports boolean indexing to filter rows based on conditions.

```python
    import pandas as pd
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['x', 'y', 'z'])
    # Select row 'y' and column 'A'
    print(df.loc['y', 'A'])
    # Select rows 'x' to 'y' (inclusive) and all columns
    print(df.loc['x':'y', :])
    # Select rows where 'A' is greater than 1
    print(df.loc[df['A'] > 1])
```

iloc (Integer-location based indexing):
Selects data based on the integer positions of rows and columns, similar to how you would index a Python list.

Can use single integers, lists of integers, or slices of integers.

When slicing with iloc, the start position is inclusive, but the end position is exclusive (like Python list slicing).

Supports boolean arrays for advanced selection.

```python
    import pandas as pd
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['x', 'y', 'z'])
    # Select the row at index 1 (second row) and column at index 0 (first column)
    print(df.iloc[1, 0])
    # Select rows from index 0 to 1 (excluding index 2) and all columns
    print(df.iloc[0:2, :])
```

### Key Differences Summarized:
loc uses labels; iloc uses integer positions.

loc slicing is inclusive of the end label; iloc slicing is exclusive of the end integer position.

loc is suitable when you know the names of your rows/columns or want to filter by conditions; iloc is suitable when you need to select by numerical position.

# Boolean Operators
You learned about the basics of Boolean logic in Python, which is essential for making decisions in your programs. Specifically, you explored how to use comparison operators to produce Boolean values (True or False) and how to combine these values using Boolean operators to perform more complex queries on your data.

Comparison Operators: You discovered that these operators compare two values and return a Boolean indicating the relationship's truthfulness. For example, x > 5 checks if x is greater than 5.
Boolean Operators: You learned about three key Boolean operators:
and: Returns True if both sides are True. For instance, True and False results in False.
or: Returns True if at least one side is True. For example, True or False results in True.
not: Negates the Boolean value it precedes. Thus, not True is False.
Combining Comparisons: You practiced combining comparison operations with Boolean operators to form more complex conditions, like checking if a variable x is between two values with x > 5 and x < 15.
Additionally, you learned how to apply these concepts to NumPy arrays and encountered the specialized functions np.logical_and, np.logical_or, and np.logical_not for element-wise logical operations. For example, to find elements of an array my_house that are greater than 18.5 or smaller than 10, you used:

```python
import numpy as np
my_house = np.array([18.0, 20.0, 10.75, 9.50])
print(np.logical_or(my_house > 18.5, my_house < 10))
```
